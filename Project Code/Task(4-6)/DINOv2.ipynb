```python
# =====================
# Cell 1: Install & Imports
# =====================
!pip -q install timm umap-learn thop tqdm

import os, torch, timm, umap, random
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader, random_split
from torchvision import datasets, transforms
from tqdm import tqdm  # Progress bar
from sklearn.metrics import classification_report
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from thop import profile

# =====================
# Cell 2: Config
# =====================
DATA_ROOT = '/kaggle/input/datasets/nahidhasan007007/group-7/PRBD Microscopic Image of Different Processed Rice/Original Images' 
IMG_SIZE = 224 
BATCH_SIZE = 64 
EPOCHS = 100    
LR = 1e-4
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'

# =====================
# Cell 3: DINOv2 Model
# =====================
class DinoV2Encoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = timm.create_model(
            'vit_base_patch14_dinov2', 
            pretrained=True, 
            num_classes=0,
            dynamic_img_size=True 
        )
        self.num_features = self.backbone.num_features
        
    def forward(self, x):
        return self.backbone(x)

# =====================
# Cell 4: Datasets
# =====================
class TransformSubset(Dataset):
    def __init__(self, subset, transform=None):
        self.subset = subset
        self.transform = transform
    def __len__(self): return len(self.subset)
    def __getitem__(self, idx):
        img, target = self.subset[idx]
        if self.transform: img = self.transform(img)
        return img, target

train_transform = transforms.Compose([
    transforms.RandomResizedCrop(IMG_SIZE),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

eval_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

full_dataset = datasets.ImageFolder(DATA_ROOT)
train_size = int(0.8 * len(full_dataset))
test_size = len(full_dataset) - train_size
train_raw, test_raw = random_split(full_dataset, [train_size, test_size])

train_ds = TransformSubset(train_raw, train_transform)
test_ds = TransformSubset(test_raw, eval_transform)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)
test_loader = DataLoader(test_ds, batch_size=BATCH_SIZE, num_workers=2)

# =====================
# Cell 5: SSL Setup
# =====================
encoder = DinoV2Encoder().to(DEVICE)
encoder.eval() 
for p in encoder.parameters(): p.requires_grad = False

proj_head = nn.Sequential(
    nn.Linear(encoder.num_features, 512),
    nn.BatchNorm1d(512),
    nn.ReLU(),
    nn.Linear(512, 128)
).to(DEVICE)

optimizer = torch.optim.AdamW(proj_head.parameters(), lr=LR)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS)

def ssl_loss(z1, z2):
    z1 = F.normalize(z1, dim=1)
    z2 = F.normalize(z2, dim=1)
    return -(z1 * z2).sum(dim=1).mean()

# =====================
# Cell 6: 100 Epoch Loop (ALL EPOCHS SHOWN)
# =====================
print(f"Starting Training on {DEVICE}...")

history = []

for epoch in range(EPOCHS):
    proj_head.train()
    total_loss = 0
    
    # tqdm creates a nice visual progress bar for each epoch
    pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}")
    
    for x, _ in pbar:
        x = x.to(DEVICE)
        v1, v2 = x, torch.flip(x, dims=[-1])
        
        with torch.no_grad():
            f1, f2 = encoder(v1), encoder(v2)
        
        z1, z2 = proj_head(f1), proj_head(f2)
        loss = ssl_loss(z1, z2)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
        pbar.set_postfix({'loss': loss.item()})
    
    scheduler.step()
    avg_epoch_loss = total_loss / len(train_loader)
    history.append(avg_epoch_loss)
    
    # This print ensures every epoch's summary is logged in the notebook output
    print(f">> Finished Epoch {epoch+1} | Avg Loss: {avg_epoch_loss:.6f} | LR: {optimizer.param_groups[0]['lr']:.8f}")

# =====================
# Cell 7: Plot Training Curve
# =====================
plt.figure(figsize=(10, 5))
plt.plot(range(1, EPOCHS+1), history)
plt.title("SSL Training Loss over 100 Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

# =====================
# Cell 8: Final Metrics
# =====================
def extract_features(loader):
    encoder.eval()
    feats, labels = [], []
    with torch.no_grad():
        for x, y in loader:
            f = encoder(x.to(DEVICE))
            feats.append(f.cpu().numpy())
            labels.append(y.numpy())
    return np.concatenate(feats), np.concatenate(labels)

X_train, y_train = extract_features(train_loader)
X_test, y_test = extract_features(test_loader)

clf = SVC(kernel='linear')
clf.fit(X_train, y_train)
print("\n--- Final Classification Report ---")
print(classification_report(y_test, clf.predict(X_test)))
```
